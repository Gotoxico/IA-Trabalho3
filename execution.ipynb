Dados de Cancer

```python
import pandas as pd
data = pd.read_csv('./archive/data.csv')
```

```python
diagnosis = data[['diagnosis']]
data = data.drop(columns=['diagnosis'])
diagnosis['diagnosis'] = diagnosis['diagnosis'].map({'M':1, 'B':0})
data = data.drop(columns=['id', 'Unnamed: 32'])
```

Separando em treino e teste

```python
separator = 0.8
train_size = int(len(data) * separator)
data_train = data[:train_size]
data_test = data[train_size:]
diagnosis_size = int(len(diagnosis)*separator)
diagnosis_train = diagnosis[:diagnosis_size]
diagnosis_test = diagnosis[diagnosis_size:]
```

Aplicando Normalização

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_train = scaler.fit_transform(data_train)
data_test = scaler.transform(data_test)
```

Importando o Modelo e treinando

```python
import LogisticRegression as RL
LogReg = RL.LogisticRegression(0.01, 1000)
LogReg.fit(data_train, diagnosis_train.values.ravel())
```

Avaliando com base nas métricas

```python
predictions = LogReg.predict(data_test)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(diagnosis_test, predictions)
precision = precision_score(diagnosis_test, predictions)
recall = recall_score(diagnosis_test, predictions)
f1 = f1_score(diagnosis_test, predictions)
print(f'Acurácia: {accuracy}')
print(f'Precisão: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
```
